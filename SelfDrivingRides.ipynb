## Self Driving Rides

```python
class Ride:
    def __init__(self, id, a, b, x, y, s, f):
        self.ID = id
        self.assigned = False
        self.start = (a,b)
        self.finish = (x,y)
        self.S = s
        self.F = f
```

```python
class Vehicle:
    def __init__(self):
        self.score = 0
        self.position = (0,0)
```

```python
class Board:
    def __init__(self, r, c, f, n, b, t, rides: list[Ride]):
        self.R = r
        self.C = c
        self.F = f
        self.N = n
        self.B = b
        self.T = t
        self.Rides = rides
```

```python
def distance(p1,p2):
    return abs(p1[0]-p2[0]) + abs(p1[1]-p2[1])

#Verify if ride starts earlier, in order to assign a bonus
def ride_start(currentstep, ride : Ride, bonus):
    if (ride.S-currentstep == 0):
        return bonus
    else:
        return 0

#Verify if the ride ends before the latest finish, in order to assign the bonus
#If the ride ends before the latest finish, the score will be equal to the distance travelled
#If not, no points are awarded
def ride_finish(currentstep, vehicle : Vehicle, ride : Ride):
    if (distance(vehicle.position, ride.finish) < currentstep):
        return (distance(ride.start, ride.finish))
    
    return 0
```

```python
#Actions

def moveleft(vehicle : Vehicle):
    vehicle.position[0] = max(0, (vehicle.position[0]-1))

def moveright(rows, vehicle : Vehicle):
    vehicle.position[0] = min(rows, (vehicle.position[0]+1))

def moveup(cols, vehicle : Vehicle):
    vehicle.position[0] = min(cols, (vehicle.position[1]+1))

def movedown(vehicle : Vehicle):
    vehicle.position[0] = max(0, (vehicle.position[1]+1))

#BASTA TER UMA VARIAVEL QUE DIGA QUANTOS STEPS E QUE O VEICULO ESTA "OCUPADO"
# NAO E NECESSARIO SABER ONDE E QUE ELE ESTA A CADA STEP
```

```python
#filename = input("Enter input file: ")
#filename = "a_example.in"
filename = "b_should_be_easy.in"

fd = open(f"input/{filename}", "r")
line = fd.readline()

R, C, F, N, B, T = map(int, line.split())

rides = []
rideid = -1
for line in fd:
    rideid += 1
    a, b, x, y, s, f = map(int, line.split())
    rides.append(Ride(rideid, a, b, x, y, s, f))
```
